## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
#The question: Which content pages should be reviewed first for refresh, given limited review capacity?
#Decision it improves: Instead of a content team guessing or manually scanning pages, this ranks pages by evidence — stale-but-visible pages, declining pages with demand, thin content getting traffic, page-one pages at risk of losing position, and low-CTR/low-engagement pages.
#Who acts on it: A content strategist or SEO reviewer with limited weekly capacity (e.g., can only review 20–50 pages per week) uses this ranked list to decide where to spend that time first.
#Cost of a wrong call: If the model ranks a low-value page too high, the reviewer wastes limited time on a page that wasn't actually a priority — meanwhile a real declining page with real traffic loss keeps losing visibility unnoticed. The cost isn't catastrophic (no page gets deleted or broken), but it's an opportunity cost: real revenue-driving pages could keep declining while attention goes to the wrong pages
```
